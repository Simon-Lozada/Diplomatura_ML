```python
# importing dataset ( example 1 and example 2 are datasets in pyECLAT)
# almacenar el conjunto de datos en una variable
dataset = Example2().get()
# printing the dataset
dataset.head()
```

```python
print(type(dataset))
```

```python
# printing the info
dataset.info()
```

```python
# cargando transacciones DataFrame a la clase ECLAT
eclat = ECLAT(data=dataset)
# DataFrame of binary values
super_bin=eclat.df_bin
print(type(super_bin))
super_bin
```

```python
eclat
```

```python
# contar elementos en cada columna
items_total = eclat.df_bin.astype(int).sum(axis=0)
items_total
```

```python
print(type(items_total))
```

```python
# contar elementos en cada fila
items_per_transaction = eclat.df_bin.astype(int).sum(axis=1)
items_per_transaction
```

```python
# Cargando elementos por estadísticas de columna en el DataFrame
df = pd.DataFrame({'items': items_total.index, 'transactions': items_total.values})
# clonación de pandas DataFrame con fines de visualización
df_table = df.sort_values("transactions", ascending=False)
#  Los 5 productos/artículos más populares
df_table.head(10).style.background_gradient(cmap='Blues')
```

```python
# tener un mismo origen
df_table["all"] = "Tree Map"
# creando un mapa de árbol usando plotly
fig = px.treemap(df_table.head(50), path=['all', "items"], values='transactions',
                  color=df_table["transactions"].head(50), hover_data=['items'],
                  color_continuous_scale='Blues',
                )
# trazando el mapa de árbol
fig.show()
```

```python
# DEMORA MÁS DE 55 MINUTOS
# el artículo debe aparecer al menos en el 5% de las transacciones
min_support = 5/100
# comenzar desde transacciones que contengan al menos 2 artículos
min_combination = 2
# hasta un máximo de artículos por transacción
max_combination = max(items_per_transaction)
rule_indices, rule_supports = eclat.fit(min_support=min_support,
                                                 min_combination=min_combination,
                                                 max_combination=max_combination,
                                                 separator=' & ',
                                                 verbose=True)
```

```python
result = pd.DataFrame(rule_supports.items(),columns=['Item', 'Support'])
result.sort_values(by=['Support'], ascending=False)
result
```
